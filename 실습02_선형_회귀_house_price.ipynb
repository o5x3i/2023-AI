#집값 데이터 읽기

```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/2023 인공지능/house_price.txt', names=['size', 'price'])
print(data)

X = data['size']
Y = data['price']
m = len(data)
```

```python
X = (np.array(X)).reshape(m,1)
Y = (np.array(Y)).reshape(m,1)
print(X.shape, Y.shape)
```

#그래프 그리기

```python
import matplotlib.pyplot as plt
plt.plot(X, Y, 'b.')
plt.xlabel("Size of a house in Pyeong")
plt.ylabel("Price in uck-won")
plt.show()
```

#Gradient descent -> theta

```python
X_b = np.c_[np.ones((m,1)), X]

learning_rate = 0.0001
n_iter = 200

theta = np.zeros((2,1))
gradients = np.zeros((2,1))

for i in range(n_iter):
  y_pred = X_b.dot(theta)
  error = y_pred - Y
  gradients = 2/m * X_b.T.dot(error)
  theta = theta - learning_rate * gradients
print("theta:")
print(theta)
```

#집 크기 최소(10평)~최대(90평) * theta -> 그래프 그리기

```python
X_new = np.array([[10], [90]])
X_new_b = np.c_[np.ones((2,1)), X_new]
Y_predict = X_new_b.dot(theta)
print("10평과 90평 집의 예측 가격")
print(Y_predict)
```

```python
plt.plot(X_new, Y_predict, 'r-')
plt.plot(X, Y, 'b.')
plt.xlabel("Size of a house in Pyeong")
plt.ylabel("Price in uck-won")
plt.show()
```

#임의의 집 크기(X_my)에 대한 집값 예측하기

```python
X_mine = np.array([[25]])
X_mine_b = np.c_[np.ones((1,1)), X_mine]
Y_predict = X_mine_b.dot(theta)
print(Y_predict, "억원")
```

#'3. Gradient descent' 대신, 확률적 경사 하강법 함수 호출하여 theta 구하기

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(max_iter = 100, penalty = None, eta0 = 0.0001)
sgd_reg.fit(X, Y.ravel())
```

```python
print("theta :")
print(sgd_reg.intercept_, sgd_reg.coef_)

Y_pred = sgd_reg.predict(X)
plt.plot(X, Y, 'b.')
plt.plot(X, Y_pred, 'r')
plt.show()
```

#비용함수 J(theta) 3차원 그래프로 그리기

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ComputeJ(t0, t1, X, Y):
  m = len(X)
  X_b = np.c_[np.ones((m,1)), X]
  theta = np.array([t0, t1])

  cost_vector = X_b.dot(theta) - Y
  cost_vector = cost_vector.reshape(-1)
  cost2 = np.square(cost_vector)
  cost_sum = np.sum(cost2) / (2 * m)

  return cost_sum
```

```python
d = 100

s = 1.0
theta0 = np.linspace(-15 * s, 15 * s, d)
theta1 = np.linspace(-1.0 * s, 1.0 * s, d)

J = np.ones((d,d))
for i, t0 in enumerate(theta0):
  for j, t1 in enumerate(theta1):
    cost = ComputeJ(t0, t1, X, Y)
    J[i, j] = cost
```

```python
pX, pY = np.meshgrid(theta0, theta1)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(pX, pY, J, 300, cmap = 'viridis')
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('J');
plt.show()
```
